```python
#Step2: Sampling strategy 
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import StratifiedKFold, cross_val_score
```

```python
# Load data and select final variables
data = data[['Attrition_Flag', 'Gender', 'Income_Category', 'Avg_Utilization_Ratio', 'Total_Revolving_Bal', 'Months_Inactive_12_mon', 'Total_Trans_Amt', 'Total_Amt_Chng_Q4_Q1']]
```

```python
# Encode 'Attrition_Flag', 'Gender', and 'Income_Category' columns
le = LabelEncoder()
data['Attrition_Flag'] = le.fit_transform(data['Attrition_Flag'])
data['Gender'] = le.fit_transform(data['Gender'])
data['Income_Category'] = le.fit_transform(data['Income_Category'])
```

```python
# Divide data into attained and existing customers
attained_customers = data[data['Attrition_Flag'] == 1]
existing_customers = data[data['Attrition_Flag'] == 0]
```

```python
# Divide attained and existing customers into 5 subsets
attained_customers_splits = np.array_split(attained_customers, 5)
existing_customers_splits = np.array_split(existing_customers, 5)
```

```python
# Create 5 balanced datasets
balanced_datasets = [pd.concat([attained_split, existing_split]) for attained_split, existing_split in zip(attained_customers_splits, existing_customers_splits)]
```

```python
#Create 5 balanced datasets
balanced_datasets = [pd.concat([attained_customers, split]) for split in existing_customers_splits] 
```

```python
#Create the combined dataset
num_existing_customers = min(1650, len(existing_customers))
selected_existing_customers = existing_customers.sample(n=num_existing_customers, random_state=42)
combined_dataset = pd.concat([attained_customers, selected_existing_customers]) 
```

```python
# Cross-validation
n_folds = 5
cv = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=42)
```

```python
# Train models on the 5 balanced datasets
for i, dataset in enumerate(balanced_datasets):
    X = dataset.drop('Attrition_Flag', axis=1)
    y = dataset['Attrition_Flag']
    
    print(f"Value counts for dataset {i+1}:")
    print(y.value_counts())
    
    clf = GaussianNB()
    clf.fit(X, y)
    
    predictions = clf.predict(X)
    print(f"Accuracy for dataset {i+1}: {accuracy_score(y, predictions)}")
    print(f"Classification report for dataset {i+1}:\n{classification_report(y, predictions)}")

```
